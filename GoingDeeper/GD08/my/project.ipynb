```python
import os
import torch
from config import PROJECT_PATH, IMAGE_SHAPE
from models import StackedHourglassNetwork, SimpleBaselinePose
from inference import predict
from visualization import draw_keypoints_on_image, draw_skeleton_on_image

```

```python
import matplotlib.pyplot as plt

num_epochs = 10  #   epoch 

# 1) SimpleBaseline: train / val    
simple_train_steps_per_epoch = len(Simple_history) // num_epochs
simple_val_steps_per_epoch   = len(Simple_val_history) // num_epochs

simple_train_epoch = [
    sum(Simple_history[i*simple_train_steps_per_epoch:(i+1)*simple_train_steps_per_epoch]) 
    / simple_train_steps_per_epoch
    for i in range(num_epochs)
]

simple_val_epoch = [
    sum(Simple_val_history[i*simple_val_steps_per_epoch:(i+1)*simple_val_steps_per_epoch]) 
    / simple_val_steps_per_epoch
    for i in range(num_epochs)
]

# 2) Hourglass: train / val    
hg_train_steps_per_epoch = len(hourglass_history) // num_epochs
hg_val_steps_per_epoch   = len(hourglass_val_history) // num_epochs

hourglass_train_epoch = [
    sum(hourglass_history[i*hg_train_steps_per_epoch:(i+1)*hg_train_steps_per_epoch]) 
    / hg_train_steps_per_epoch
    for i in range(num_epochs)
]

hourglass_val_epoch = [
    sum(hourglass_val_history[i*hg_val_steps_per_epoch:(i+1)*hg_val_steps_per_epoch]) 
    / hg_val_steps_per_epoch
    for i in range(num_epochs)
]

# 3) Hourglass loss 4   
hourglass_train_epoch_div4 = [x / 4.0 for x in hourglass_train_epoch]
hourglass_val_epoch_div4   = [x / 4.0 for x in hourglass_val_epoch]

# 4)  
epochs = range(1, num_epochs + 1)

plt.figure(figsize=(12, 5))

# ------- Train Loss -------
plt.subplot(1, 2, 1)
plt.plot(epochs, simple_train_epoch, label='SimpleBaseline train', marker='o')
plt.plot(epochs, hourglass_train_epoch, label='Hourglass train (sum)', marker='s')
plt.plot(epochs, hourglass_train_epoch_div4, label='Hourglass train (/4)', marker='^')
plt.xlabel('Epoch')
plt.ylabel('Train Loss')
plt.title('Train Loss per Epoch')
plt.grid(True, alpha=0.3)
plt.legend()

# ------- Val Loss -------
plt.subplot(1, 2, 2)
plt.plot(epochs, simple_val_epoch, label='SimpleBaseline val', marker='o')
plt.plot(epochs, hourglass_val_epoch, label='Hourglass val (sum)', marker='s')
plt.plot(epochs, hourglass_val_epoch_div4, label='Hourglass val (/4)', marker='^')
plt.xlabel('Epoch')
plt.ylabel('Val Loss')
plt.title('Validation Loss per Epoch')
plt.grid(True, alpha=0.3)
plt.legend()

plt.tight_layout()
plt.show()
```
